```python
import pandas as pdhttps://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.DataFrame.plot.line.html
df = pd.DataFrame({
   'pig': [20, 18, 489, 675, 1776],
   'horse': [4, 25, 281, 600, 1900]
   }, index=[1990, 1997, 2003, 2009, 2014])
lines = df.plot.line()
```

```python
axes = df.plot.line(subplots=True)
type(axes) # <class 'numpy.ndarray'>
#axes[0:10]
```

```python
df = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
ax = df.plot.barh(x='lab', y='val')
```

```python
df = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
ax = df.plot.bar(x='lab', y='val', rot=0) # rotate x and y
```

```python
import saeaborn as sns
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```
