```python
import pandas as pd, numpy as np
```

```python
clust_attb = pd.read_csv('/dbfs/FileStore/tables/DATA_SCIENCE/Cluster_Attribute_Profile_500K_0629.csv')
```

```python
clust_attb['Whole Grain Percent:range'].iloc[0]
```

```python
cat_attbs = [i for i in clust_attb.columns if 'unique' in i]
num_attbs = [i for i in clust_attb.columns if 'range' in i]
```

### Generate list of non-null attributes per cluster

```python
notnull_dict = {}
for i in range(len(clust_attb)): # 7159
    row_ = clust_attb.iloc[i]
    notnull = []
    for col in cat_attbs:
        if row_[col] != '[nan]': notnull.append(col.split(':')[0])
    for col in num_attbs:
        if row_[col] != '(nan, nan)': notnull.append(col.split(':')[0])
    notnull_dict[i] = notnull
```

```python
clust_attb['Non-Null Attributes'] = clust_attb.ClusterLabel.map(notnull_dict)
```

```python
clust_attb[['ClusterLabel','Cluster Name', 'Non-Null Attributes']].to_csv('/dbfs/FileStore/tables/DATA_SCIENCE/Cluster_NonNullAttributes_500K.csv', index = None)
```
