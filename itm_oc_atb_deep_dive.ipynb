# ITM_OC_ATB_deep_dive
 - Deep dive the relationship between Attributes_Index_Readable_Headers_060923.csv (Attributes_Index), created from CURIDQ.PIMMART.DIM_ATB_INF_LKP (DIM), and CURIDQ.PIMMAR.ITM_OC_ATB (ITM)
   1. Create a list of column names specified in  Attributes_Index with 'Table in PIM-Mart' = 'ITM_OC_ATB'
      1.1.  Remove column names which do not exist in ITM_OC_ATB   
   2. Invoke function unq_val to create a DataFrame with value count for columns from 1.1 and save it to CSV file 
 - Author: Sophia Yue
 - History
   - Jun 27, 23: Initialized 

```python
def unq_val(df, l_col):
    """
    Purpose:
      - Create dataframe to include  
        - df0: keep feature name 
        - df1: get count for unique values from the feature
        - df2: get ratio of count for unique values from the feature 
        - df3: Merge df1 and df2 
      - Concating dfy, df0, df3 with axis = 0 
    Parameter
     input 
      df : df to be analyzed 
      l_col:  list of features
     output 
      dfy : dataframe for the count of unique values and  ratio of unique values    
    """
    dfy =  pd.DataFrame() 
    l_comment=[]
    for fea in l_col:   
       
          
           df0 = pd.DataFrame([[fea,'','', ' ']],columns=['feature','value','cnt','ratio_cnt'] )
           df1 = df[fea].value_counts(dropna = False).to_frame(name = 'cnt').reset_index()
           df1.rename(columns={'index':'value'}, inplace = True)
           sd =  df[fea].value_counts(dropna = False)/len(df)
           df2 = sd.to_frame(name='ratio_cnt').reset_index()
           df2.rename(columns={'index':'value'}, inplace = True)
           df3 = df1.merge(df2, on = 'value',how = 'inner')
           dfy = pd.concat([dfy,  df0, df3], axis = 0)


    return dfy  
   
```

## Get value count for all the column names in attributes_Index_Readable_Headers_060923.csv

```python
path_file_atb = path + 'Attributes_Index_Readable_Headers_060923.csv'
df_atb = pd.read_csv(path_file_atb)
```

```python
l_col_itm_oc_atb = list(df_atb[df_atb['Table in PIM-Mart'] == 'ITM_OC_ATB']['Attribute Value Column Name']) 
```

```python
l_rmv =['FLD_TO_OVD_VLU', 'KRG_IDCd_FIRE_PVNN_SGE_FL', 'OVD_VLU','PLUM_DPT_NO','SUP_ID_FOR_OVD_VLU']
l_col = [x for x in  l_col_itm_oc_atb if x not  in l_rmv]  
l_col = l_col + ['KRG_IDC_FIRE_PVNN_SGE_FL', 'PLUM_DPT_VLU']
df_dim_item = unq_val(df, l_col )
df_dim_item['feature'] = df_dim_item['feature'].replace(np.NaN, '')
df_dim_item['value'] = df_dim_item['value'].replace(np.NaN, 'missing')
```
