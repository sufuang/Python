```python
df_set_list = df_flavor_atb.groupby('COM_CD', dropna = False).agg({'comb_flavor_cod_val': lambda x: list(set(x))})
df_set_list.reset_index(inplace=True)
df_set_list
```

```python

df_gtin_cnt = df_flavor_atb.groupby('COM_CD', dropna = False)['GTIN_NO'].apply(lambda x: x.isna().count()).reset_index(name ='GTIN_CNT')
df_gtin_cnt
```

```python
df_gtin_cnt=  pd.merge(df_gtin_cnt, df_set_list,  on ='COM_CD', how = 'inner')
df_gtin_cnt['no_of_flavor'] =  df_gtin_cnt['comb_flavor_cod_val'].apply(lambda x: len(x))
df_gtin_cnt = df_gtin_cnt  [['COM_CD','GTIN_CNT','no_of_flavor', 'comb_flavor_cod_val']]
df_gtin_cnt
```
