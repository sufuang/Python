```python
def node_depth(Node_ID, hierarchy = wine_hierarchy):
    hierarchy['Topics_Count'] = hierarchy.Topics.apply(len)
    hierarchy.sort_values('Topics_Count', ascending= False, inplace= True)
    count = 0
    while(True):
        ancestor = hierarchy[(hierarchy.Child_Left_ID == Node_ID) |(hierarchy.Child_Right_ID == Node_ID)]
        if len(ancestor) == 0: break
        Node_ID = ancestor.Parent_ID.values[0]
        count += 1
    return count
```

```python
wine_hierarchy['Node_Depth'] = wine_hierarchy.Parent_ID.apply(lambda x: node_depth(Node_ID = x))
```
