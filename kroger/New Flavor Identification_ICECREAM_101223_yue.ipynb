```python
import re
def extract_bnj_flavor(desc):
    pattern = r'[0-9]' #Split on number
    splitted = re.split(pattern, desc.upper())[0]
    if 'ICE CREAM' in splitted: return splitted.split('ICE CREAM')[1]
    else: return splitted 
```
