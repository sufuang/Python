## Get column values  for ‘acds_transactions.xlsx’ and  'gtin_agg.xls' 

```python
def get_col_val(df, cut_off = 50):
    l_val = []
    for fea in df.columns.values:
        l_fea_uniq =  list(df[fea].unique())
        type = df.dtypes[fea]  # get datatype
        len_uniq = len(l_fea_uniq)
        rpt_all_unique_ind = 'Yes' 
        if len(l_fea_uniq) > cut_off:
            rpt_all_unique_ind ='No'       
            l_fea_uniq = l_fea_uniq[0:cut_off]

        l_val.append([fea, type, len_uniq , rpt_all_unique_ind, l_fea_uniq])
    df_col_val = pd.DataFrame(l_val,columns=['Attribute Value Column Name','Type', 'Unique Length ', 'Report all unique', 'col_val'])  
    return    df_col_val
```
